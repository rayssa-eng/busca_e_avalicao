```python
import configparser
import csv
```

### Lendo arquivo de configuração

```python
config = configparser.ConfigParser()
config.read('BUSCA.cfg')

modelo_path = config.get('DEFAULT', 'MODELO')

modelo_vetorial = {}
with open(modelo_path, 'r', newline='') as file:
    reader = csv.reader(file, delimiter=';')
    
    # Iterate over each row in the CSV file
    for row in reader:
        key = row[0].strip()
        value = eval(row[1].strip())
        modelo_vetorial[key] = value

consultas_path = config.get('DEFAULT', 'CONSULTAS')
results_path = config.get('DEFAULT', 'RESULTADOS')
```
